이제부터 메일을 보내게 될 텐데, 아래와 같은 형식으로 보내되, 이메일과 비밀번호를 자신의 것으로 바꿔야 하고, 보내다가 이메일 형식이 이상한 경우 에러가 뜰 수 있다. 이때 이름과 번호가 같이 있는 사람은 이미 보내진 것이므로, 제외해야 한다. 

```python
title = "[소속] 홍보 기사 작성을 부탁드립니다"
yag = yagmail.SMTP("my_mail@gmail.com","password1234")


for i in range(len(names)):
    print(i,end=' ')
    body = '''
    안녕하세요, %s %s 기자님
    
    저는 누구입니다..
    
    
    본 내용....
    
    날씨가 쌀쌀한데, 언제나 건강하시고 무탈하시길 기원합니다.
    

    감사합니다.
    누구 드림
    
    
    ========================================
    직책
    이름

    M) 보낸사람 전화번호: 010-1234-1234
    E) 보낸사람 메일 주소 : somebody@somemail.com
    B) 애드쿠스 홈페이지: page.com
    F) 애드쿠스 인스타: @instagramID

    '''% (news[i], names[i])
    
    yag.send(
        to=mails[i],
        subject=title,
        contents=body, 
    )
    print(names[i])
```
